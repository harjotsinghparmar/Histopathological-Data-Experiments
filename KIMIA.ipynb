## Take glimpse of all classes

```
cwd = "/content/kdataset"
#create a list of class folders
class_folders = os.listdir(cwd)
#use list comprehension to sample each class
sample_each_class = [(cwd+'/'+sample+'/'+ os.listdir(cwd+'/'+sample)[0]) for sample in class_folders]
print(sample_each_class)
```

### Use matplotlib to tile all the 20 classes in 4*5 tiles 

```
cols,rows = 4,5
fig = plt.figure()
for i in range(1,cols*rows+1):
  a = fig.add_subplot(cols,rows,i)
  image = cv2.imread(sample_each_class[i-1])
  plt.imshow(image)
fig.suptitle('Class Samples')

```
